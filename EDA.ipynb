01. Column Name: StarReview
    Type of Graph: Box Plot
    Description: The individuals who can buy mobiles, will give a stars based on the performance of the product. The highest star is 5 and here we have                   different star reviews on each mobile.
                 The line in the middle of a box plot represents the median of the data. The median is the middle value of a dataset when it is arranged                  in ascending order. It divides the data into two equal halves:
                 --> 50% of the data points lie below the median.
                 --> 50% of the data points lie above the median.

```python
### Bivariate Analysis 
```

Bivariate Analysis will be done on using multiple columns where we can easily understand the graphs and analyze the information in the data. 

Column Names: Brand, StarReview. 
Graph Type: Bar
Description: The below data is graph between Continuous Vs Categorical where, continuous means the numerical data and Categorical means string data. 
I have plotted a bar graph using groupby where it takes Brand column as a categorical data and StarReview column as a Continuous. The graph shows us the Starreviews of each brand. 

```python
plt.figure(figsize = (8,6))
grouped_data = df.groupby('Brand')['StarReview'].mean().reset_index()
sns.barplot(x='Brand', y='StarReview', data=grouped_data, palette = 'Set2')
plt.title('Groupby: Continuous vs Categorical')
plt.show()
```

Column Names: DisplaySize (cm), Version, Brand. 
Graph Type: Pivot Table 
Description: The rows is the table having the types of mobile versions, the columns contains brand names and the values in the table are the mean value of the specific version and brand of the mobile. 

```python
pivot_data = df.pivot_table(values='DisplaySize (cm)', 
                            index='Version', 
                            columns='Brand', 
                            aggfunc='mean')

sns.heatmap(pivot_data, annot=True, cmap='Blues')
plt.title('Pivot Table: Continuous vs Categorical')
plt.show()
```

Description: A heatmap is a graphical representation of data where individual values are represented as colors in a matrix or grid. It is commonly used to visualize correlations, distributions, and patterns within a dataset.

Column Names: StarReview, Reviews, RAM (GB)
Graph Type: Corelation Heatmap 
Description: The rows and columns represent categories, variables, or indices of the data.
Color Intensity: The intensity of the color reflects the magnitude of the data point.
For example:
--> Darker colors might represent higher values.
--> Lighter colors might represent lower values.

```python
# Correlation matrix
correlation_matrix = df[['StarReview', 'Reviews', 'RAM (GB)']].corr()

# Heatmap for correlation
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Plot: Continuous vs Continuous')
plt.show()
```

A crosstab (cross-tabulation) is a table that summarizes the relationship between two or more categorical variables. It displays the frequency (or count) of occurrences for combinations of values from the variables.
The values in a crosstab are typically frequencies or aggregated values, such as:
Counts: The number of occurrences for each combination of categories.
Percentages: Proportions or percentages of the total dataset for each combination.
Aggregations: If a numerical variable is involved, the crosstab can show sums, averages, or other statistics for each combination of categories.

```python
# Crosstab between two categorical variables
crosstab_data = pd.crosstab(df['Brand'], df['Version'])

# Heatmap for crosstab data
sns.heatmap(crosstab_data, annot=True, cmap='Blues', fmt='d')
plt.title('Crosstab: Categorical vs Categorical')
plt.show()
```
