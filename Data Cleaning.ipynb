Our Dataframe consists missing values and those should be filled with particular values to visualize the data. 

For Numerical columns we can fill the nan values with mean, median or mode of that particular column. 
-> The Rating column is filled with the mean value of the rating column, it will first calculate the mean of the avaliable values then fill the none values with that mean value. 

-> The Reviews column is filled with the median value of the Reviews column, it will first calculate the median of the avaliable values then fill the none values with that median value. 

-> The RAM (GB) column is filled with the mode value of the RAM (GB) column, it will first calculate the mode of the avaliable values which is the most repeated value in the column and fill it with that mode value. 

For Categorical columns we can fill the missing values only with Mode value of that specific column. 
-> The Version column is filled with the mode value of the Version column, it will first calculate the mode of the avaliable values then fill the none values with that mode value. 

-> The Color column is filled with the mode value of the Color column, it will first calculate the mode of the avaliable values then fill the none values with that mode value. 

```python
df['Rating'].fillna(df['Rating'].mean(), inplace=True)  
df['Reviews'].fillna(df['Reviews'].median(), inplace=True)  
df['RAM (GB)'].fillna(df['RAM (GB)'].mode()[0], inplace=True)  

df['Version'].fillna(df['Version'].mode()[0], inplace=True)  
df['Color'].fillna(df['Color'].mode()[0], inplace=True) 
```

```python
df['StarReview'].fillna(df['StarReview'].median(), inplace=True)  
```
