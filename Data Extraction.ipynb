Till now we have extracted the data from a single URL which gives limited amount of data which is not sufficient for our project. The below code takes the information from multiple pages where we mention the range for it, the libraries which have been used will read the complete content of the websites. After that we can extract the data like, costs, reviews, brand, version etc. and append it into our list, missing values will be appended as Nan. 

```python
ProductName = []
ReviewByStar = [] 
SalesPrice = [] 
ActualPrice = [] 
Features = [] 
Ratings = [] 
request_header={'Content-Type':'text/html;charset=UTF-8','User-Agent':'Chrome/101.0.0.0 (Windows NT 10.0,Win64;x64;rv:109.0) Gecko/20100101 Firefox/119.0','Accept-Encoding':'gzip, deflate,br'}

for i in range(1,21):
    URL = 'https://www.flipkart.com/search?q=mobiles&as=on&as-show=on&otracker=AS_Query_HistoryAutoSuggest_1_5_na_na_na&otracker1=AS_Query_HistoryAutoSuggest_1_5_na_na_na&as-pos=1&as-type=HISTORY&suggestionId=mobiles&requestId=a472c5e7-380d-4571-bf64-6b338ad693f5&page=2'.format(i)
    page = requests.get(URL, headers = request_header)
    pagecontent = page.text
    soup = BeautifulSoup(pagecontent)

    for x in soup.find_all('div', attrs = {'class': 'yKfJKb row'}):
    
        Price = x.find('div', attrs={'class': 'Nx9bqj _4b5DiR'})
        MobileName = x.find('div', attrs={'class': 'KzDlHZ'})
        Star = x.find('div', attrs = {'class': 'XQDdHH'})
        RealPrice = x.find('div', attrs = {'class': 'yRaY8j ZYYwLA'})
        Feature = x.find('ul', attrs = {'class': 'G4BRas'})
        Rating = x.find('span', attrs = {'class': 'Wphh3N'})

        if Price is None:
            SalesPrice.append(np.NaN)
        else:
            SalesPrice.append(Price.text)

        if MobileName is None:
            ProductName.append(np.NaN)
        else:
            ProductName.append(MobileName.text)

        if Star is None:
            ReviewByStar.append(np.NaN)
        else:
            ReviewByStar.append(Star.text)

        if RealPrice is None:
            ActualPrice.append(np.NaN)
        else:
            ActualPrice.append(RealPrice.text)

        if Feature is None:
            Features.append(np.NaN)
        else:
            Features.append(Feature.text)

        if Rating is None:
            Ratings.append(np.NaN)
        else:
            Ratings.append(Rating.text)
```

The above code will retrive every data we mentioned in the code, then we need to frame it in a specific manner where we can easily understand the data. 

```python
df = pd.DataFrame({
    'ProductName': ProductName,
    'SalesPrice': SalesPrice,
    'ActualPrice': ActualPrice,
    'StarReview': ReviewByStar,
    'Ratings': Ratings,
    'Features': Features
})

df
```

From here we are going to generate new columns from existed columns. 

The below regex expression helps us to Extract only the brand values ('Apple', 'Samsung', 'Motorola', etc.) from the ProductName column. 

```python
regex = "^[^ ]+"
df['Brand'] = df['ProductName'].apply(lambda x: re.findall(regex, x)[0])
```

The following regex expression i used will help to extract the Verison values. Ex: 'POCO C61' from this product it will give 'C61' as the output in the version column. 

```python
regex = r"(?:(?<!\w)\d{3}[^\s,]*\s(?:Single SIM|5G|4G)|T3(?:x| Pro| Plus| Pro 5G)|Galaxy\s[A-Z]+\s[^\s,]*|iPhone\s[^\s]+|C\d+\b|13C\b|Nord\sCE\s\d+\sLite\s5G)"
df['Version'] = df['ProductName'].apply(lambda x: re.search(regex, x).group(0) if re.search(regex, x) else np.nan)
```

The below regular expression will give only the color from the column ProductName. Ex: 'REDMI 13C (Starfrost White, 128 GB)' from this it will give 'Starfrost White' and appended into the new column named as color. 

```python
regex = r'\(([^,]+)\b'
df['Color'] = df['ProductName'].apply(lambda x: re.search(regex, x).group(1) if re.search(regex, x) else np.nan)
```

From the features column, extracting the size of RAM where regrex expression describes that, when the RAM value is in MB it will be divided by 1000 to convert it into GB and when it is TB it can be multiplied by 1000. 

```python
regex = r'(\d+)\s*(MB|GB|TB)\s*RAM'

def extract_ram_value(text):
    if isinstance(text, str): 
        match = re.search(regex, text)
        if match:
            number = int(match.group(1))  # Extract number
            unit = match.group(2)  # Extract unit (MB, GB, TB)
            if unit == 'MB':
                return number / 1000  # Convert MB to GB
            elif unit == 'TB':
                return number * 1000  # Convert TB to GB
            return number  # Return GB value as is
    return np.nan  # Return NaN if no match or not a string

df['RAM (GB)'] = df['Features'].apply(extract_ram_value)
```

```python
df 
```

Same as i mentioned above. 

```python
regex = r'(\d+)\s(GB|MB|TB)\sROM'

def extract_rom_convert(x):
    if isinstance(x, str):
        match = re.search(regex, x)
        if match:
            number = int(match.group(1))
            unit = match.group(2)
            if unit == 'MB':
                return float(number) / 1000  # Convert MB to GB by dividing by 1000, return as float
            elif unit == 'TB':
                return float(number) * 1000  # Convert TB to GB by multiplying by 1000, return as float
            else:
                return float(number)  # Return the number for GB as float
    return np.nan  # Return NaN if no match or it's not a string

df['ROM (GB)'] = df['Features'].apply(extract_rom_convert)
```
